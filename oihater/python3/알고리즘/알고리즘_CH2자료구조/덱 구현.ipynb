```python
import sys
from collections import deque
deq = deque()
length = int(input()) # input()쓰면 시간초과 input()대신 sys.stdin.readline()
for i in range(length):
    cmd = input().split()
    if cmd[0] == 'push_back':
        deq.append(cmd[1])
    elif cmd[0] == 'push_front':
        deq.appendleft(cmd[1])
    elif cmd[0] == 'front':
        if len(deq) == 0:
            print(-1)
        else:
            print(deq[0])
    elif cmd[0] == 'back':
        if len(deq) == 0:
            print(-1)
        else:
            print(deq[-1])
    elif cmd[0] == 'size':
        print(len(deq))
    elif cmd[0] == 'pop_front':
        if len(deq) == 0:
            print(-1)
        else:
            print(deq.popleft())
    elif cmd[0] == 'pop_back':
        if len(deq) == 0:
            print(-1)
        else:
            print(deq.pop())
    elif cmd[0] == 'empty':
        if len(deq) == 0:
            print(1)
        else:
            print(0)
```

```python
#class로구현
import sys
from collections import deque
deq = deque()
length = int(sys.stdin.readline()) # input을 쓰면 시간초과
for i in range(length):
    cmd = sys.stdin.readline().split()
    if cmd[0] == 'push_back':
        deq.append(cmd[1])
    elif cmd[0] == 'push_front':
        deq.appendleft(cmd[1])
    elif cmd[0] == 'front':
        print(deq[0])
    elif cmd[0] == 'back':
        print(deq[-1])
    elif cmd[0] == 'size':
        print(deq.size())
    elif cmd[0] == 'pop_front':
        if len(deq) == 0:
            print(-1)
        else:
            print(deq.popleft())
    elif cmd[0] == 'pop_back':
        if len(deq) == 0:
            print(-1)
        else:
            print(deq.pop())
    elif cmd[0] == 'empty':
        if len(deq) == 0:
            print(1)
        else:
            print(0)
```
